```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
langs = ['fr','es','en']
emb_types = ['doc2vec', 'bert']
df = pd.DataFrame()
for lang in langs:
    for emb_type in reversed(emb_types):
        df = df.append(pd.read_csv('./../../results/Classification/results' + lang + emb_type + '.csv', index_col=0))
```

```python
df.set_index(np.linspace(0, 5, num = 6, dtype=np.int), inplace=True)
df
```

```python
df.groupby('lang').mean().plot.bar(figsize=(12,8))
plt.grid()
plt.title('Text classification language comparison')
plt.xlabel('Languages')
plt.ylabel('Percentaje')
plt.savefig('./../../results/Classification/lang_comparison.png')
```

```python
df.groupby('embedding').mean().plot.bar(figsize=(12,8))
plt.grid()
plt.title('Text classification embedding comparison')
plt.xlabel('Embedding')
plt.ylabel('Percentaje')
plt.savefig('./../../results/Classification/embedding.png')
```

```python
import os, json
import pandas as pd
import numpy as np
```

```python
csv_files = sorted(os.listdir('./data/News/'))
counter = len(os.listdir('/media/juan/Juan/NLP/FakeNews/Dataset/en/'))
print(counter)
```

```python
counter = len(os.listdir('/media/juan/Juan/NLP/FakeNews/Dataset/en/'))
for file in csv_files:
    if 'Fake' not in file:
        continue
    df = pd.read_csv('./data/News/' + file)
    for index in range(0, df.index[-1]):
        news = df.loc[index, :]
        news_dict = {'source': 'COVID-News Dataset',
                     'url': news['news_url'],
                     'title': news['title'],
                     'lang': 'en'}
        if 'content' in news:
            news_dict['text'] = news['content']
        if 'publish_date' in news:
            news_dict['date'] = news['publish_date']
        save_file = open('/media/juan/Juan/NLP/FakeNews/Dataset/en/fake_' + str(counter) + '.json', 'w+')
        print(save_file)
        json.dump(news_dict, save_file, indent=4)
        save_file.close()
        counter += 1
```

```python
import matplotlib.pyplot as plt
```

```python
langs = ['en', 'fr', 'es']
sources = ['twitter', 'reddit', 'news']
tweets = [651761, 260000, 125000]
reddit = [297736, 4682, 9150]
news = [447754, 699669, 257754]
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
metrics_df = pd.read_csv('./../../results/FakeNewsDetection/fake_news_detection_results.csv', index_col=0)
metrics_df.drop(metrics_df.columns[0], inplace=True, axis=1)
metrics_df.head()
```

```python
metrics_df.groupby('model').mean().plot.bar(figsize = (12, 8))
plt.title('Comparaci칩n del desempe침o de diferentes modelos', fontsize=20)
plt.ylabel('Porcentaje', fontsize=15)
plt.xlabel('Modelo', fontsize=15)
plt.grid(which='both')
plt.savefig('./../../results/FakeNewsDetection/model_comparison.png')
```

```python
metrics_df.groupby('lang').mean().plot.bar(figsize = (12, 8))
plt.title('Comparaci칩n del desempe침o en diferentes idiomas', fontsize=20)
plt.ylabel('Porcentaje', fontsize=15)
plt.xlabel('Idioma', fontsize=15)
plt.grid(which='both')
plt.savefig('./../../results/FakeNewsDetection/language_comparison.png')
```

```python
metrics_df.groupby('features').mean().plot.bar(figsize = (12, 8))
plt.title('Comparaci칩n del desempe침o de extracci칩n de caracter칤sicas', fontsize=20)
plt.ylabel('Porcentaje', fontsize=15)
plt.xlabel('Extracci칩n de caracter칤sticas', fontsize=15)
plt.grid(which='both')
plt.savefig('./../../results/FakeNewsDetection/feature_extraction.png')
```
